```python
# Adding values to the existing DataFrame
df['Grade']=['A','B','C','D']
df
```

```python
# Using loc() to add row or column/access/display row or column
#loc() helps you find and get specific parts of your table using names or labels, like finding a treasure in a map by pointing to its location!
# df.loc[:] - Consider entire row
df.loc[:,'Grade2']=['A','B','C','D']
df
```

```python

df.loc[4]=['Nancy',5,'A','B']
df
```

```python
#Renmaing the index for the new added entry
df.rename(index={4: 'E'}, inplace=True)
print(df)
```

```python
# Removing duplicates
df=df.drop_duplicates()
df
```

```python
# Remove a column

#axis=1: This tells pandas that you want to drop a column (not a row).
#inplace=True: This will modify the DataFrame in place. If you don't use inplace=True, the method will return a new DataFrame with the column removed.

df.drop('Grade2',axis=1,inplace=True)
df
```

```python
#Remove a row
#axis=0: This tells pandas that you want to drop a column (not a row).
df.drop('E',axis=0,inplace=True)
df
```

```python
#Return a tuple representing the dimensions of the DataFrame.
df.shape
```

```python
#Dropping index A and B
df.drop(['A','B'], axis=0,inplace=True)
df
```

```python
#Rename a column
df.rename(columns={'ID':'Student ID'},inplace=True)
df
```

```python
#Rename a row
df.rename(index={'D':'E'},inplace=True)
df
```
